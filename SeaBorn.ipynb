```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Create Plot Using Random DataSet.

```python
Var1 = [1,2,3,4,5]
Var2 = [11,22,33,44,55]

data_set = pd.DataFrame({"A": Var1, "B" : Var2})

# Print Data_Set
data_set.head()
```

```python
sns.lineplot(x = Var1, y = Var2, data = data_set)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Line Chart")
plt.show()
```

# Import DataSet from the External Source

```python
Data = sns.load_dataset('penguins')
Data.head(5)
```

# Line Plot

```python
Chart = sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = Data, hue = 'sex')
Chart.set_xlabel("Bill Length")
Chart.set_ylabel("Bill Depth")
Chart.set_title('Line Plot')
plt.show()
```

# Histogram Plot

```python
Chart1 = sns.histplot(x = 'flipper_length_mm', data = Data, hue = 'sex',binwidth = 3,kde = True)
Chart1.set_title('Hist Plot')
plt.show()
```

```python
Chart2 = sns.histplot( y = 'flipper_length_mm', data = Data, hue = 'sex', color = ['green','red'], binwidth = 3,multiple = "stack")
Chart2.set_title('Hist Plot')
plt.show()
```

```python
Chart3 = sns.displot( x = 'flipper_length_mm', data = Data, hue = 'sex', binwidth = 3, kde = True)
plt.title('Hist Plot')
plt.show()
```

```python
Chart3 = sns.displot( y = 'flipper_length_mm', data = Data, hue = 'species', binwidth = 3, multiple = "stack")
plt.title('Hist Plot')
plt.show()
```

```python
# Some Others Atteributes

# bins = Number
# kde = True/False
# Multiple = stack/dodge
# element = step/poly
# log_scale = True/False
# Fill = Flase/True
```

# Bar Chart

```python
Var1 = [10,24,33,47,52,45,56,79,34,80]
Var2 = [11,22,33,44,55,23,54,85,49,50]
Data = pd.DataFrame({"A": Var1, "B": Var2 })

sns.barplot(data = Data, x = Var1, y = Var2, color = 'red', saturation = 0.9)
sns.set(style = "darkgrid")
plt.title("Bar Chart")
plt.show()
```

# Scatter Plot

```python
Chart1 = sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data = Data, hue = 'sex', style = "sex", palette = 'deep')
plt.xlabel("Bill Length")
plt.ylabel("Bill Depth")
plt.title("Scatter Plot")
plt.show()
```

# Count Plot

```python
# Load DataSet From the Local Machine

Data = pd.read_csv("tips.csv")
Data.head(2)
```

# X - Axis Count Plot

```python
sns.countplot(x = 'sex', hue = 'smoker', data = Data, width = 0.5, gap = 0.1 )
plt.title('Count Plot')
plt.show()
```

 # Y - Axis Count Plot

```python
sns.countplot(y = 'sex', hue = 'smoker', data = Data, width = 0.5, gap = 0.1 )
plt.title('Count Plot')
plt.show()
```

# Violin Plot

```python
cls = "green"
sns.violinplot(x = 'day', y = 'total_bill', data =Data, linewidth = 2 ,color = cls)
plt.title("Violin Plot")
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data =Data, hue = "sex", palette = 'Dark2')
plt.title("Violin Plot")
plt.show()
```

# Pair Plot

```python
# Show Data_Set

Data.head(3)
```

```python
sns.pairplot(data = Data,  vars = ['tip', 'total_bill', 'size'], hue = 'sex')
plt.show()

# Atteributes
# dig_kinds = hist
# kind = hist/reg/kde/satter
```

 # Strip Plot

```python
sns.stripplot(x = "day", y = 'total_bill', data = Data, hue = "sex", size = 5, linewidth = 1)
plt.title('Strip Plot')
plt.show()
```

```python
sns.stripplot(y = Data["total_bill"], data = Data, hue = "smoker", size = 5, linewidth = 1)
plt.title('Strip Plot')
plt.show()
```

# Box Plot

```python
# X - Axis Box Plot

sns.boxplot(x = 'day', y = 'total_bill', data = Data, color = 'g', linewidth = 2)
plt.title('Box Plot')
plt.show()
```

```python
# X - Axis Box Plot

sns.boxplot(y = 'day', x = 'total_bill', data = Data, color = 'g', linewidth = 2)
plt.title('Box Plot')
plt.show()
```
